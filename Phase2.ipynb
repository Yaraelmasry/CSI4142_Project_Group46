```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset using the raw version of the dataset using our github repository that contains the datasets
## fetch the datasets from this repo https://github.com/Yaraelmasry/CSI4142_Project_Group46

url = "https://raw.githubusercontent.com/Yaraelmasry/CSI4142_Project_Datasets/main/Datasets/vgsales.csv"
```

```python
# loading our First Dataset (Video Game Sales)
print (url)
vgsales = pd.read_csv(url)
```

```python
vgsales.head()
vgsales.describe()
```

###  Cleaning the Data


```python
# check for missing values
vgsales.isnull().sum()

# drop rows with missing values
vgsales = vgsales.dropna()

# check for duplicate rows
vgsales.duplicated().sum()

# drop duplicate rows
vgsales = vgsales.drop_duplicates()

# check for invalid data types
vgsales.dtypes

# convert data types if necessary
vgsales['Year'] = vgsales['Year'].astype('int64')

# check for outliers
vgsales.describe()

# handle outliers if necessary
vgsales
```

### Adding New Features to the Dataset

```python

# new feature for the total sales of each game
vgsales['Total_Sales'] = vgsales['NA_Sales'] + vgsales['EU_Sales'] + vgsales['JP_Sales'] + vgsales['Other_Sales']

# new feature for the average sales of each game
vgsales['Average_Sales_Game'] = vgsales['Total_Sales'] / vgsales['Global_Sales']

# new feature for the sales per platform
vgsales['Sales_Per_Platform'] = vgsales['Total_Sales'] / vgsales['Platform'].nunique()

# new feature for the average sales per year
vgsales['Average_Sales_Per_Year'] = vgsales['Global_Sales'] / vgsales['Year'].nunique()
vgsales
```

# Dataset 3
# Game Backlogs, Plays, Playing

```python
# Loading the dataset using the raw version of the dataset using our github repository that contains the datasets
## fetch the datasets from this repo https://github.com/Yaraelmasry/CSI4142_Project_Datasets

backlogs_url = "https://raw.githubusercontent.com/Yaraelmasry/CSI4142_Project_Datasets/main/Datasets/backloggd_games.csv"
```

```python
print (backlogs_url)
backlogs = pd.read_csv(backlogs_url)
```

```python
backlogs
```

```python
# remove lists column and Summary
backlogs.drop(columns=['Lists'], inplace=True)
backlogs.drop(columns=['Summary'], inplace=True)
```

## Cleaning the Data:
- Missing Values
- Outliers
- Typos
- Surrogate Key
- Duplicate Rows
- Invalid DataTypes



```python
# checks for missing values
backlogs.isnull().sum()

# drops rows with missing values
backlogs = backlogs.dropna()

# checks for duplicate rows
backlogs.duplicated().sum()

# drops duplicate rows
backlogs = backlogs.drop_duplicates()

# checks for invalid data types
backlogs.dtypes

# converts data types if necessary
backlogs['Rating'] = backlogs['Rating'].astype('float')

# checks for outliers
backlogs.describe()

# handles outliers if necessary
backlogs['Rating'] = backlogs['Rating'].clip(0, 10)

# remove typos
backlogs['Platforms'] = backlogs['Platforms'].str.replace('PC (Microsoft Windows)', 'PC')
backlogs['Platforms'] = backlogs['Platforms'].str.replace('PlayStation 4', 'PS4')
backlogs['Platforms'] = backlogs['Platforms'].str.replace('Xbox One', 'XOne')

# generate surrogate keys
backlogs['SK_Game'] = backlogs.index

backlogs
```

### Removes 'Unnamed' column and replaces it with Surrogate Key of the name

```python
if 'Unnamed: 0' in backlogs.columns:
    backlogs.drop(columns=['Unnamed: 0'], inplace=True)

backlogs.rename(columns={'game_id': 'SK_Game'}, inplace=True)
# reordering so that SK_Game Appears
columns = backlogs.columns.tolist()
columns = ['SK_Game'] + [col for col in columns if col != 'SK_Game']
backlogs = backlogs[columns]
backlogs
```

### Feature Engineering
- Extracting Release Year and Calculating Game Age
- Splitting the release data column to year, month, day
- Adding a new feature of Game Age

```python
# extract Release Year
backlogs['Release_Date'] = backlogs['Release_Date'].replace('TBD', pd.NaT)

# convert the Release_Date column to datetime
backlogs['Release_Date'] = pd.to_datetime(backlogs['Release_Date'])

# split the Release_Date into year, month, and day
backlogs['Release Year'] = backlogs['Release_Date'].dt.year.astype('Int64')

# Calculating Game Age
backlogs['Game Age'] = 2024 - backlogs['Release Year']

backlogs
```

```python
# convert to string
backlogs['Plays'] = backlogs['Plays'].astype(str)
backlogs['Playing'] = backlogs['Playing'].astype(str)
backlogs['Backlogs'] = backlogs['Backlogs'].astype(str)
backlogs['Wishlist'] = backlogs['Wishlist'].astype(str)

# remove 'K' suffix and convert to numeric
backlogs['Plays'] = backlogs['Plays'].str.replace('K', 'e3').astype(float)
backlogs['Playing'] = backlogs['Playing'].str.replace('K', 'e3').astype(float)
backlogs['Backlogs'] = backlogs['Backlogs'].str.replace('K', 'e3').astype(float)
backlogs['Wishlist'] = backlogs['Wishlist'].str.replace('K', 'e3').astype(float)

# calculate total games owned
backlogs['Total Games Owned'] = backlogs['Playing'] + backlogs['Backlogs'] + backlogs['Wishlist']

# calculate plays to backlogs ratio
backlogs['Plays to Backlogs Ratio'] = backlogs['Plays'] / backlogs['Backlogs']

# calculate playing ratio
backlogs['Playing Ratio'] = backlogs['Playing'] / backlogs['Total Games Owned']

# replace infinite value with NaN
backlogs.replace([np.inf, -np.inf], np.nan, inplace=True)

# drop Nan values
backlogs.dropna(axis=0, inplace=True)

backlogs
```
